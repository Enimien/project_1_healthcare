```python
# Import dependancies.
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import linregress
import time
import warnings
warnings.filterwarnings("ignore")
```

```python
# Import data file.
daily_activities = "Fitabase Data 4.12.16-5.12.16\dailyActivity_merged.csv"
```

```python
# Read data file.
daily_activities_df = pd.read_csv(daily_activities)
```

```python
# Display data file.
daily_activities_df
```

```python
# Check the data type for each column.
daily_activities_df.info()
```

```python
# Check for missing values.
missing_values = daily_activities_df.isnull().sum()
missing_values
```

```python
# Convert date to datetime.
daily_activities_df["ActivityDate"] = pd.to_datetime(daily_activities_df["ActivityDate"])
daily_activities_df.head()
```

```python
# Sum the total active minutes and active distance add it to the daily_activities dataframe in new columns.
daily_activities_df['TotalActiveMinutes'] = (daily_activities_df['VeryActiveMinutes'] +
                                             daily_activities_df['FairlyActiveMinutes'] +
                                             daily_activities_df['LightlyActiveMinutes'])


daily_activities_df
```

```python
# Convert active time to hours.
daily_activities_df["ActiveHours"] = round(daily_activities_df["TotalActiveMinutes"] / 60, 2 )

daily_activities_df
```

```python
# Change the dates to Days of the Week
daily_activities_df["DayOfTheWeek"] = daily_activities_df["ActivityDate"].dt.day_name()
daily_activities_df
```

```python
# Drop unwanted Columns
daily_activities_df = daily_activities_df.drop(columns=["TrackerDistance", "LoggedActivitiesDistance", "SedentaryActiveDistance", "SedentaryMinutes"])
daily_activities_df
```

```python
# Get the summary of the data.
daily_activities_df.describe()
```

```python
# Find the total number of users.
user_count = daily_activities_df["Id"].nunique()
print(f"The total number of users are {user_count}.")
```

```python
# Find hour at which users are most active.
hourly_activity = daily_activities_df.groupby("ActiveHours")["TotalSteps"].mean().reset_index()
hourly_activity["ActiveHours"] = hourly_activity["ActiveHours"].astype(int)

plt.figure(figsize=(10, 6))
sns.barplot(data=hourly_activity, x="ActiveHours", y="TotalSteps")
plt.title("Average Steps per Hour")
plt.xlabel("Hour of the Day")
plt.ylabel("Average Steps")
plt.show()
```

```python
# Find the averages of the user data and store it in a Dataframe
avg_steps_per_user = daily_activities_df.groupby(["Id"])["TotalSteps"].mean()
avg_distance_per_user = daily_activities_df.groupby(["Id"])["TotalDistance"].mean()
veryactv_mins_per_user = daily_activities_df.groupby(["Id"])["VeryActiveMinutes"].mean()
fairlyactv_mins_per_user = daily_activities_df.groupby(["Id"])["FairlyActiveMinutes"].mean()
lightlyactv_mins_per_user = daily_activities_df.groupby(["Id"])["LightlyActiveMinutes"].mean()
avg_active_mins = daily_activities_df.groupby(["Id"])["TotalActiveMinutes"].mean()
avg_calories_per_user = daily_activities_df.groupby(["Id"])["Calories"].mean()

user_summary = pd.DataFrame({
                        "Average Steps Per User": avg_steps_per_user,
                        "Average Distance Per User": avg_distance_per_user,
                        "Average Very Actv Mins Per User": veryactv_mins_per_user,
                        "Average Fairly Actv Mins Per User": fairlyactv_mins_per_user,
                        "Average Lightly Actv Mins Per User": lightlyactv_mins_per_user,
                        "Average Active Mins Per User": avg_active_mins,
                        "Average Calories Burnt Per User": avg_calories_per_user
                        })
user_summary.head()
```

```python
# Create a correlation matrix for the various variables.
correlation_matrix = user_summary.corr()

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.xticks(rotation= 45, ha="right")
plt.title("Correlation Matrix of Fitness Data")
plt.tight_layout()
plt.show()
```

```python
# Create a scatter plot for the average steps and average calories per user.
plt.figure(figsize=(10, 6))
sns.scatterplot(data=user_summary, x="Average Steps Per User", y="Average Calories Burnt Per User", color="b", alpha=0.7)
plt.title("Average Calories Burnt per User per Average Steps")
plt.xlabel("Average Steps per User")
plt.ylabel("Average Calories Burnt per User")
plt.grid(True)
plt.show()
```

```python
# Create a scatter plot for the average distance and average calories burnt per user.
plt.figure(figsize=(10, 6))
sns.scatterplot(data=user_summary, x="Average Distance Per User", y="Average Calories Burnt Per User", color="g", alpha=0.7)
plt.title("Average Calories Burnt per User per Average Distance")
plt.xlabel("Average Distance per User")
plt.ylabel("Average Calories Burnt per User")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=user_summary, x="Average Very Actv Mins Per User", y="Average Calories Burnt Per User", color="g", alpha=0.7)
plt.title("Average Calories Burnt per User per Average Very Actv Mins")
plt.xlabel("Average Very Actv Mins Per User")
plt.ylabel("Average Calories Burnt per User")
plt.grid(True)
plt.show()
```

```python
# Generate lineplot of the total steps and total calories burnt for a single user with time.
user_id = 2022484408
user_activity = daily_activities_df[daily_activities_df["Id"] == user_id]


fig, ax1 = plt.subplots(figsize=(10, 6))
plt.xticks(rotation= 45)

sns.lineplot(data=user_activity, x="ActivityDate", y="TotalSteps", marker="o", ax=ax1, color="g", label="TotalSteps", alpha=0.6, legend=False)
ax1.set_ylabel("TotalSteps", color="g")
ax1.tick_params(axis="y", labelcolor="g")

ax2 = ax1.twinx()
sns.lineplot(data=user_activity, x="ActivityDate", y="Calories", marker="o", ax=ax2, color="b", label="Calories", alpha=0.6, legend=False)
ax2.set_ylabel("Calories", color="b")
ax2.tick_params(axis="y", labelcolor="b")

plt.title(f"Steps and Calories Burned Per Day for User {user_id}")
ax1.set_xlabel("Date")

lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc="upper left")
plt.show()
```

```python
# Perfoem a linear regression on the avarage steps and calories burnt
avg_steps = user_summary["Average Steps Per User"]
avg_calories = user_summary["Average Calories Burnt Per User"]
slope_steps, intercept_steps, r_value_steps, p_value_steps, std_err_steps = linregress(avg_steps, avg_calories)
line_eqn = f" y = {slope_steps: .2f} x  + {intercept_steps: .2f})"

plt.figure(figsize=(10, 6))
sns.scatterplot(x=avg_steps, y=avg_calories, label='Data Points', color = "g", legend=False)
plt.plot(avg_steps, intercept_steps + slope_steps * avg_steps, color='red', label='Regression Line')
plt.annotate(line_eqn,(min(avg_steps)+10, min(avg_calories)+1500), fontsize=15, color="red")
plt.title('Average Steps vs Average Calories')
plt.xlabel('Average Steps')
plt.ylabel('Average Calories')
plt.show()
```
